# Reading a CSV with no Header

```python
# Without a header in the CSV, Pandas will use the first row as the header
salesnon_df = pd.read_csv(
    csvpath,
    header=None
)
```

```python
# Read the file without a header
salesnon_df
```

```python
# Rewrite the column names
col= [
        'Full Name',
        'Email',
        'Address',
        'Zip Code',
        'Credit Card',
        'Sales Price'
]
```

```python
salesnon_df.columns = col
salesnon_df.head()
```

# Rename Function:
> __*rename (_)*__

> __*Mapping*__ \- use Python dictionary to map the old column name to the
new one: __*"Old Column Name": "New Column Name"*__

```python
salesyes_df.rename(
    columns={
        'FullName': 'Full Name',
        'CreditCard': 'Credit Card',
        'SalePrice': 'Sale Price'
    }
)
```
